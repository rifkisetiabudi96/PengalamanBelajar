# **Cleaning Data**

```python
# Remove duplicated data for customers_df
customers_df.drop_duplicates(inplace = True)
```

```python
# Check to make sure data is not duplicated again
customers_df.duplicated().sum()
```

```python
# Check missing value
customers_df[customers_df.gender.isna()]
```

```python
# Check dominan value for gender column
customers_df.gender.value_counts()
```

Prefer no to say is the most value in gender column, it is the mode

```python
# Fill gender missing value with mode
customers_df.fillna(value = "Prefer not to say", inplace=True)
# Check again dominan value for gender column
customers_df.gender.value_counts()
```

Solve inaccurate value

```python
# Check maximum for age column
customers_df[customers_df.age == customers_df.age.max()]
```

I make assumption the is an typo for the input data, it should be 70, not 700

```python
# Replace 700 to 70
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
# Make sure maximum for age column
customers_df[customers_df.age == customers_df.age.max()]
```

There is a same problem, I assume the same thing, then replace 500 to 50

```python
# Replace 500 to 50
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
# Check again for the maximum value
customers_df[customers_df.age == customers_df.age.max()]
```

Chage Data type for order_date and delivery date

```python
# Make datetime column
datetime_column = ["order_date", "delivery_date"]
for column in datetime_column:
  orders_df[column] = pd.to_datetime(orders_df[column])

# check dtype for orders_df
orders_df.info()
```

Remove duplicated values for product_df

```python
# Remove
product_df.drop_duplicates(inplace=True)

# Check duplicated value
print("Sum of duplicated:", product_df.duplicated().sum())
```

Solve missing value for sales_df

```python
# Check missing value
sales_df[sales_df.total_price.isna()]
```

```python
# Replace NaN to price_per_unit product with quantity
sales_df['total_price'] = sales_df['price_per_unit'] * sales_df['quantity']

# Check again missing value
sales_df.isna().sum()
```

Convert Dataframe to csv

```python
customers_df.to_csv("customers.csv")
orders_df.to_csv("orders.csv")
sales_df.to_csv("sales.csv")
product_df.to_csv("product.csv")
```
